```python
import requests

# 定义目标接口的URL
url = 'http://www.xianxiangchain.xyz/DAPP/login.php?'
# 用户账号、密码和APP安全码
user_account = 'Shufan Cai'
user_password = '1234'
app_security_code = 'DAPP2023'#

# 构建POST请求的数据
data = {
    'userAccount': user_account,
    'userPassword': user_password,
    'appSecurityCode': app_security_code
}

# 发送POST请求
response = requests.post(url, data=data)

# 检查响应状态码
if response.status_code == 200:
    # 登录成功
    print('Login successful')
    print(response.json())  # 输出服务器返回的JSON数据
elif response.status_code == 401:
    # 登录失败
    print('Login failed. Check your credentials.')
    print(response.json())  # 输出服务器返回的错误信息
elif response.status_code == 405:
    # 不允许的请求方法
    print('Method Not Allowed')
    print(response.json())  # 输出服务器返回的错误信息
else:
    # 其他错误
    print(f'Request failed with status code {response.status_code}')
    print(response.text)  # 输出服务器返回的文本内容
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties

font_path = './GenYoMin-B.ttc'

data1 = np.array([4.44, 22.22])
data2 = np.array([95.46, 77.78])

categories = ['观察组', '对照组']


width = 0.3
plt.figure(figsize=(8, 6))

font_prop = FontProperties(fname=font_path)

plt.bar(categories, data1, color='red', alpha=1.0, label='发生率', width=width)
plt.bar(categories, data2, color='blue', alpha=1.0, label='未发生率', bottom=data1, width=width)

plt.ylabel('Precentage', fontproperties=font_prop,fontsize=15)
plt.title('两组并发症发生率对比', fontproperties=font_prop,fontsize=15)

plt.legend(prop=font_prop, loc='upper right')

plt.rcParams['font.sans-serif'] = ['SimSun']  # SimSun is a font that includes a wide range of Chinese characters

# Show the plot
plt.show()


```
